## **ISING MODEL WITHOUT CHECKERBOARD UPDATE**

```python
import matplotlib.pyplot as plt
import numpy as np
from numba import njit #For Faster Computation
```

## Snapshots of after reaching equilibrium

```python
def Ising_system(N): #To create random spin system
    spin = np.random.choice([-1, 1], (N, N))
    return spin
```

```python
@njit
def update_spin(Spin, T, color):
    L = len(Spin)

    for i in range(L):
        for j in range(L):
            if (i + j) % 2 == color: #Either odd/even positioned spin update

                s = Spin[i, j]

                nn = (
                    Spin[(i + 1)%L, j] +
                    Spin[(i - 1)%L, j] +
                    Spin[i, (j + 1)%L] +
                    Spin[i, (j - 1)%L]
                )

                delta_E = 2 * nn * s        #energy difference while spin flipping

                if delta_E < 0 or np.random.random() < np.exp(-delta_E/T):
                    Spin[i, j] = -s

    return Spin
```

```python
T_val = [1, 1.5, 1.75, 2, 2.25, 2.5]
pl = 0
plt.figure(figsize = (15, 10))
for T in T_val:
    spin = Ising_system(500)
    for i in range(200):
        spin = update_spin(spin, T, 0)
        spin = update_spin(spin, T, 1)
    pl += 1
    plt.subplot(2, 3, pl)
    plt.imshow(spin, cmap = 'gray')
    plt.title(f"Ising system for T = {T}", size = 10)
    plt.axis('off')
plt.show()
```

## **Average Magnetization**

```python
@njit
def Magnetization(spin, N_eq, N_iter, T):
    for i in range(N_eq):
        spin = update_spin(spin, T, 0) #Update spin to get it into equilibrium
        spin = update_spin(spin, T, 1)
    M_av = np.empty(N_iter)
    M2_av = np.empty(N_iter)
    for j in range(N_iter):
        spin = update_spin(spin, T, 0)
        spin = update_spin(spin, T, 1)
        M_av[j] = abs(np.mean(spin))
        M2_av[j] = abs(np.mean(spin))**2
    return np.mean(M_av), np.mean(M2_av) #|<M>| and <M**2>
```

```python
N = 20
Temperature = np.linspace(1, 4, 30)
M_T = np.empty(30)
var = 0
for Temp in Temperature:
    spin = Ising_system(N)
    M_av = Magnetization(spin, 400, 3000, Temp)[0] #|<M>|
    M_T[var] = M_av
    var += 1
```

```python
plt.plot(Temperature, M_T, marker = 'o')
plt.grid()
plt.xlabel("Temperature")
plt.axvline(2.269, ls='--', color='gray', label=r'$T_c$')
plt.ylabel("<|M|>")
plt.show()
```

## **Average Energy**

```python
@njit
def En(spin):
    energy = 0
    N = len(spin[0])
    for i in range(N):
        for j in range(N):
            s = spin[i, j]
            nn = (spin[(i + 1)%N, j] + spin[(i - 1)%N, j] + spin[i, (j + 1)%N] + spin[i, (j - 1)%N])
            energy += -s*nn
    return energy/(4*N*N) #Average Interaction Energy
```

```python
@njit
def Average_En(spin, N_eq, N_iter, T):
    for i in range(N_eq):
        spin = update_spin(spin, T, 0) #Update spin to get it into equilibrium
        spin = update_spin(spin, T, 1)
    E_av = np.empty(N_iter)
    E2_av = np.empty(N_iter)
    for j in range(N_iter):
        spin = update_spin(spin, T, 0)
        spin = update_spin(spin, T, 1)
        E_av[j] = En(spin)
        E2_av[j] = En(spin)*En(spin)
    return np.mean(E_av), np.mean(E2_av)
```

```python
N = 20
Temperature = np.linspace(1, 4, 30)
E_T = []
for Temp in Temperature:
    spin = Ising_system(N)
    E_av = Average_En(spin, 300, 40000, Temp)[0]
    E_T.append(E_av)
```

```python
plt.plot(Temperature, E_T, marker = 'o')
plt.grid()
plt.xlabel("Temperature")
plt.axvline(2.269, ls='--', color='gray', label=r'$T_c$')
plt.ylabel("<E>")
plt.show()
```

## **Specific Heat**

```python
C_T = []
for Temp in Temperature:
    spin = Ising_system(N)
    E, E2 = Average_En(spin, 300, 40000, Temp)
    C_T.append((E2 - E**2)/Temp**2)           #C
```

```python
plt.plot(Temperature, C_T, marker = 'o')
plt.grid()
plt.xlabel("Temperature")
plt.axvline(2.269, ls='--', color='gray', label=r'$T_c$')
plt.ylabel("Specific Heat")
plt.show()
```

We know that, $<E^2> - <E>^2 = K_BT^2C_v$; here I took $K_B$ = 1

```python
X_T = []
for Temp in Temperature:
    spin = Ising_system(N)
    M, M2 = Magnetization(spin, 300, 20000, Temp)
    X_T.append((M2 - M**2)/Temp)
```

```python
plt.scatter(Temperature, X_T)
plt.grid()
plt.xlabel("Temperature")
plt.axvline(2.269, ls='--', color='gray', label=r'$T_c$')
plt.ylabel("Susceptibility")
plt.show()
```

We know that, $<M^2> - <M>^2 = K_BTX$ where X is susceptibilty; here I took $K_B$ = 1
